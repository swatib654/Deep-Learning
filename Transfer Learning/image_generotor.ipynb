```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
# Create  an image data generator with augementations
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.4,
    height_shift_range=0.4,
    shear_range=0.4,
    zoom_range=0.4,
    horizontal_flip=True,
    rescale=None,
    fill_mode='reflect'
)

# Load the imge
img = load_img(r"C:\Users\pc\OneDrive\Desktop\Deep Learning\Transfer Learning\images.jpeg")
```

```python
img
```

```python
# convert to numpy array
x = img_to_array(img)
x = x.reshape((1,) + x.shape)

# generate 
i=0
for batch in datagen.flow(x, batch_size=1,
                          save_to_dir=r"C:\Users\pc\OneDrive\Desktop\Deep Learning\Transfer Learning\image_create",
                          save_prefix='Happy image', save_format='jpeg'):
    i += 1
    if i > 30:
        break
    
```
